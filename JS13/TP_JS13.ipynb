Load dataset MNIST dari Keras.

Bangun model dengan 2 hidden layer.

Latih model dan evaluasi akurasi.

```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical
import time

# 1. Load dataset MNIST
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Normalisasi data (0-255 → 0-1)
X_train = X_train / 255.0
X_test = X_test / 255.0

# One-hot encoding label
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# 2. Bangun model JST
model = Sequential([
    Flatten(input_shape=(28, 28)), # Ubah gambar 28x28 menjadi vektor
    Dense(256, activation='relu'), # Hidden layer 1 dengan 256 neuron
    Dense(128, activation='relu'), # Hidden layer 2 dengan 128 neuron
    Dense(10, activation='softmax') # Output layer (10 kelas)
])

# 3. Kompilasi model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# 4. Latih model
print("\nTraining Model...")
start_time = time.time()
model.fit(X_train, y_train, epochs=10, batch_size=32, verbose=1)
end_time = time.time()
training_time = end_time - start_time
print(f"Training selesai dalam {training_time:.2f} detik")

# 5. Evaluasi model
print("\nEvaluating Model...")
loss, acc = model.evaluate(X_test, y_test, verbose=0)
print(f"Akurasi pada data uji: {acc:.4f}")
```


Ubah jumlah neuron di hidden layer (misal: 256 dan 128).

Tambahkan satu hidden layer lagi.

Bandingkan akurasi dan waktu pelatihan.

```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical
import time

# 1. Load dataset MNIST (sama seperti sebelumnya)
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Normalisasi data (0-255 → 0-1)
X_train = X_train / 255.0
X_test = X_test / 255.0

# One-hot encoding label
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# 2. Bangun model JST
print("Membangun Model dengan 3 Hidden Layers (512, 256, 128 neuron)...")
model_modified = Sequential([
    Flatten(input_shape=(28, 28)), # Ubah gambar 28x28 menjadi vektor
    Dense(512, activation='relu'), # Hidden layer 1: 512 neuron (Diubah)
    Dense(256, activation='relu'), # Hidden layer 2: 256 neuron (Diubah)
    Dense(128, activation='relu'), # Hidden layer 3: 128 neuron (Ditambahkan)
    Dense(10, activation='softmax') # Output layer (10 kelas)
])

# 3. Kompilasi model
model_modified.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# 4. Latih model
print("\nTraining Model Modifikasi...")
start_time = time.time()
# Melatih dengan epochs=10, batch_size=32 (sama seperti model asli)
model_modified.fit(X_train, y_train, epochs=10, batch_size=32, verbose=1)
end_time = time.time()
training_time_modified = end_time - start_time
print(f"Training selesai dalam {training_time_modified:.2f} detik")

# 5. Evaluasi model
print("\nEvaluating Model Modifikasi...")
loss_modified, acc_modified = model_modified.evaluate(X_test, y_test, verbose=0)
print(f"Akurasi Model Modifikasi pada data uji: {acc_modified:.4f}")
```

Model modifikasi dengan tiga hidden layer (512, 256, 128 neuron) mencapai akurasi uji 0.9817 setelah dilatih selama 214.26 detik dan mendapatkan training loss akhir 0.0164.


Eksperimen dengan fungsi aktivasi Sigmoid vs ReLU.

```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical
import time

# 1. Load dataset MNIST (sama seperti sebelumnya)
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Normalisasi data (0-255 → 0-1)
X_train = X_train / 255.0
X_test = X_test / 255.0

# One-hot encoding label
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# 2. Bangun model JST dengan Aktivasi SIGMOID
print("Membangun Model dengan 3 Hidden Layers (512, 256, 128 neuron) menggunakan SIGMOID...")
model_sigmoid = Sequential([
    Flatten(input_shape=(28, 28)), # Ubah gambar 28x28 menjadi vektor
    Dense(512, activation='sigmoid'), # Diubah: ReLU -> Sigmoid
    Dense(256, activation='sigmoid'), # Diubah: ReLU -> Sigmoid
    Dense(128, activation='sigmoid'), # Diubah: ReLU -> Sigmoid
    Dense(10, activation='softmax') # Output layer (tetap softmax)
])

# 3. Kompilasi model
model_sigmoid.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# 4. Latih model
print("\nTraining Model dengan SIGMOID...")
start_time = time.time()
# Melatih dengan epochs=10, batch_size=32
model_sigmoid.fit(X_train, y_train, epochs=10, batch_size=32, verbose=1)
end_time = time.time()
training_time_sigmoid = end_time - start_time
print(f"Training selesai dalam {training_time_sigmoid:.2f} detik")

# 5. Evaluasi model
print("\nEvaluating Model dengan SIGMOID...")
loss_sigmoid, acc_sigmoid = model_sigmoid.evaluate(X_test, y_test, verbose=0)
print(f"Akurasi Model Sigmoid pada data uji: {acc_sigmoid:.4f}")
```
